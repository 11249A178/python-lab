```python
from openpyxl import Workbook

# Create a new workbook
wb = Workbook()
ws = wb.active

# Write data into cells
ws['A1'] = "Name"
ws['B1'] = "Age"

ws['A2'] = "John"
ws['B2'] = 25

ws['A3'] = "Emma"
ws['B3'] = 30

# Save the file
wb.save("data.xlsx")

print("Data written to data.xlsx successfully!")
```
