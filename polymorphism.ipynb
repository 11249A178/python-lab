```python
# Base class
class PalindromeChecker:
    def is_palindrome(self, value):
        pass     # This will be overridden


# Derived class for Strings
class StringPalindrome(PalindromeChecker):
    def is_palindrome(self, value):
        value = value.lower().replace(" ", "")
        return value == value[::-1]


# Derived class for Integers
class IntegerPalindrome(PalindromeChecker):
    def is_palindrome(self, value):
        value_str = str(value)
        return value_str == value_str[::-1]


# -------- Main Program --------

# Input from user
inp = input("Enter a string or integer: ")

# Check whether input is integer or string
if inp.isdigit():
    obj = IntegerPalindrome()   # polymorphism: same method, different class
    result = obj.is_palindrome(int(inp))
else:
    obj = StringPalindrome()
    result = obj.is_palindrome(inp)

if result:
    print("Palindrome")
else:
    print("Not a Palindrome")
```
