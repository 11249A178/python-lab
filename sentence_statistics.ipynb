```python
# Store the problem description in a string
Q5 = """5. Sentence Statistics:
Write a Python program that accepts a sentence and finds the number of words, digits, and uppercase letters.
"""

# Print the problem description
print(Q5)

# Input sentence from user
sentence = input("Enter a sentence: ")

# Count words, digits, and uppercase letters
num_words = len(sentence.split())
num_digits = sum(char.isdigit() for char in sentence)
num_uppercase = sum(char.isupper() for char in sentence)

# Display results
print(f"Number of words: {num_words}")
print(f"Number of digits: {num_digits}")
print(f"Number of uppercase letters: {num_uppercase}")
```
