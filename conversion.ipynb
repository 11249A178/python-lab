```python
Q4 = """Binary to Decimal & Octal to Hexadecimal Conversion:
Develop a Python program to convert binary to decimal, and octal to hexadecimal.
"""

# Print the problem description
print(Q4)

# Binary to Decimal conversion
binary_num = input("Enter a binary number: ")
decimal_num = int(binary_num, 2)
print(f"Decimal equivalent of binary {binary_num} is {decimal_num}")

# Octal to Hexadecimal conversion
octal_num = input("Enter an octal number: ")
decimal_from_octal = int(octal_num, 8)
hexadecimal_num = hex(decimal_from_octal)[2:].upper()
print(f"Hexadecimal equivalent of octal {octal_num} is {hexadecimal_num}")
```
