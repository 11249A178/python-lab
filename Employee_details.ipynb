```python
class Employee:
    def __init__(self, name, emp_id, department, salary):
        self.name = name
        self.emp_id = emp_id
        self.department = department
        self.salary = salary

    def display(self):
        print(f"Name: {self.name}, ID: {self.emp_id}, Dept: {self.department}, Salary: {self.salary}")

    # Method to update salary
    def update_salary(self, amount):
        self.salary += amount


# ---- Main Program ----

# List to store employee objects
employees = [
    Employee("John", 101, "IT", 45000),
    Employee("Emily", 102, "HR", 40000),
    Employee("David", 103, "IT", 50000),
    Employee("Sophia", 104, "Finance", 55000)
]

department_to_update = "IT"
increment_amount = 5000

print("\n--- Before Salary Update ---")
for emp in employees:
    emp.display()

# Update salary for employees in given department
for emp in employees:
    if emp.department == department_to_update:
        emp.update_salary(increment_amount)

print("\n--- After Salary Update ---")
for emp in employees:
    emp.display()
```
