```python
# Store the problem statement
Q8 = """File Operations: Write a Python program to accept a file name from the user and
perform the following operations:
1. Display the first N lines of the file
2. Find the frequency of occurrence of a word accepted from the user in the file
"""
print(Q8)

# Accept file name from the user
filename = input("Enter the file name (with extension): ")

try:
    # Open the file in read mode
    with open(filename, 'r') as file:
        lines = file.readlines()

    # 1. Display the first N lines
    N = int(input("Enter the number of lines to display: "))
    print(f"\nFirst {N} lines of the file:")
    for line in lines[:N]:
        print(line, end='')

    # 2. Find frequency of a word
    word = input("\n\nEnter the word to find its frequency: ")
    content = ' '.join(lines)
    word_count = content.lower().split().count(word.lower())
    print(f"\nThe word '{word}' occurs {word_count} times in the file.")

except FileNotFoundError:
    print("Error: File not found. Please check the file name and try again.")
```
